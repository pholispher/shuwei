```python
import pandas as pd

data = pd.read_excel('e1r1.xlsx')

# Display the first few rows of the data to understand its structure
data.head()

# Clean the data by removing rows with missing data in key columns
data_cleaned = data.dropna(subset=['时间', '正己烷不溶物（INS)g', '焦油产率', '水产率', '焦渣产率'])

# Calculate the correlation between INS and the various yields
correlation_tar_ins = data_cleaned['焦油产率'].corr(data_cleaned['正己烷不溶物（INS)g'])
correlation_water_ins = data_cleaned['水产率'].corr(data_cleaned['正己烷不溶物（INS)g'])
correlation_char_ins = data_cleaned['焦渣产率'].corr(data_cleaned['正己烷不溶物（INS)g'])

# Import necessary visualization library
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style of visualization
sns.set(style="whitegrid")

# Create a figure to plot the relationships
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Plot each yield correlation with INS
sns.lineplot(ax=axes[0], x=data_cleaned['正己烷不溶物（INS)g'], y=data_cleaned['焦油产率'], color='blue')
axes[0].set_title(f'Correlation between INS and Tar Yield: {correlation_tar_ins:.2f}')
axes[0].set_xlabel('N-Hexane Insoluble (g)')
axes[0].set_ylabel('Tar Yield')

sns.lineplot(ax=axes[1], x=data_cleaned['正己烷不溶物（INS)g'], y=data_cleaned['水产率'], color='green')
axes[1].set_title(f'Correlation between INS and Water Yield: {correlation_water_ins:.2f}')
axes[1].set_xlabel('N-Hexane Insoluble (g)')
axes[1].set_ylabel('Water Yield')

sns.lineplot(ax=axes[2], x=data_cleaned['正己烷不溶物（INS)g'], y=data_cleaned['焦渣产率'], color='red')
axes[2].set_title(f'Correlation between INS and Char Yield: {correlation_char_ins:.2f}')
axes[2].set_xlabel('N-Hexane Insoluble (g)')
axes[2].set_ylabel('Char Yield')

plt.tight_layout()
plt.show()
```
